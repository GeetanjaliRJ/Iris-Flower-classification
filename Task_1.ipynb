```python
df = df.drop('Id', axis = 1)
df
```

Visualization

```python
sns.pairplot(df, hue = 'Species', palette = 'hls')
```

```python
x = df.drop(['Species'],1)
x.head()
```

```python
y = df['Species']
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2)
```

 Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression 
```

```python
lr_model = LogisticRegression()
lr_model.fit(x_train, y_train)
```

```python
Predection1 = lr_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, Predection1)*100)
```

Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model_dtc = DecisionTreeClassifier()
model_dtc.fit(x_train, y_train)
```

```python
Predection2 = model_dtc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, Predection2))
0.9
```

 Support Vector Machine Algorithm

```python
from sklearn.svm import SVC
model_svc = SVC()
```

```python
model_svc.fit(x_train, y_train)
```

```python
Predection3 = model_svc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,Predection3)*100)
```

From Logistics regression and Support Vector Machine Algorithm model I got 96% accuracy and from DecisionTreeClassifier() I got 90% accuracy
